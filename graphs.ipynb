```python
import matplotlib.pyplot as plt

# Data for the three splits
# Note: Membership originally had 5% and 85% (sum 90%). Assuming a typo, I set Sector B to 95%.
membership = {'Sector A': 5, 'Sector B': 95}
claims = {'Claims A': 42, 'Claims B': 58}
bills = {'Bills A': 9, 'Bills B': 91}

# Create a figure with 1 row and 3 columns
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
fig.suptitle('Splits Overview', fontsize=16)

# Colors for A and B (consistent across charts)
colors = ['#ff9999', '#66b3ff']  # light red and light blue

# Function to add a pie chart
def make_pie(ax, data, title):
    wedges, texts, autotexts = ax.pie(
        data.values(),
        labels=data.keys(),
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        textprops={'fontsize': 12}
    )
    # Make percentage text bold and white for better contrast
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
    ax.set_title(title, fontsize=14)

# Plot each split
make_pie(axes[0], membership, 'Membership Split')
make_pie(axes[1], claims, 'Claims Split')
make_pie(axes[2], bills, 'Bills Split')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
